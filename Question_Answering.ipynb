#2. Load Model and Tokenizer

 loading the pre-trained model (deepset/roberta-base-squad2) and its tokenizer using AutoModelForQuestionAnswering and AutoTokenizer. This model is fine-tuned on the SQuAD2 dataset.

```python
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
model_name = "deepset/roberta-base-squad2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

# 3. Test the model

let's check the model performance

```python
context = (
    "Hugging Face is a leading company in the field of artificial intelligence. "
    "It specializes in natural language processing and offers the popular Transformers library, "
    "which provides access to numerous pre-trained models for tasks like text classification, "
    "translation, summarization, and question answering. Additionally, Hugging Face hosts a hub "
    "for sharing models and datasets among the research community."
)
```

```python
questions = [
    "What does Hugging Face specialize in?",
    "What library is Hugging Face known for?",
    "What kind of tasks can you perform with their models?",
    "What does the Hugging Face Hub offer?"
]
```

```python
for question in questions:
    inputs = tokenizer(question, context, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    start_index = torch.argmax(outputs.start_logits)
    end_index = torch.argmax(outputs.end_logits) + 1
    answer = tokenizer.convert_tokens_to_string(
        tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start_index:end_index])
    )
    print("Question:", question)
    print("Answer:", answer)
    print("-" * 50)
```
